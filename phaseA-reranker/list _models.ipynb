```python
import os

base_dir = "/data/bioasq13/phaseA-reranker/trained_models_b02"

dirs2 = []
num_models = 0
for root, dirs, _ in os.walk(base_dir):
    for d in sorted(dirs):
        if d.startswith("checkpoint-"):
            num_models += 1
            full_path = os.path.join(root, d)
            rel_path = os.path.relpath(full_path, base_dir)
            # print('"'+rel_path+'"')
            dirs2.append(rel_path)

for i in sorted(dirs2):
    print('"${MODEL_DIR}/' + i + '"')
print(num_models)
```

```python
import os

base_dir = "/data/bioasq13/outputs/Batch04/runs_a/2023"

num_models = 0
for root, dirs, files in os.walk(base_dir):
    for f in sorted(files):
        if "pointwise" in f and "relevance" not in f:
            continue
        num_models += 1
        full_path = os.path.join(root, f)
        rel_path = os.path.relpath(full_path, base_dir)
        print('"${RUN_DIR}/2023/' + rel_path + '"')
print(num_models)
```
